```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.columns
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head(4)
```

```python
## To check data type 
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_name = final_df.Country.value_counts().index
country_name
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
## pie chart - Top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observation : Zomato maximum records or transaction are from India after that USA and then United Kingdom

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation 
1. When rating are between 4.5 to 4.9 --> Excellent
2. When rating are between 4.0 to 4.4 --> Very Good
3. When rating are between 3.5 to 3.9 --> Good
4. When rating are between 2.5 to 3.4 --> Average
5. When rating are betwwen 1.8 to 2.4 --> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, hue='Rating color', palette=['gray','red','orange','yellow','green','darkgreen'])
```

Observaton : 
1. Not rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
 ## Count plot

sns.countplot(x='Rating color', data=ratings, hue='Rating color', palette=['gray','red','orange','yellow','green','darkgreen'])
```
